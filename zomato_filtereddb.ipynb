```python
import pandas as pd
from sqlalchemy import create_engine
import time

# 1. High-Performance CSV Load
file_path = r'C:\Users\swati\Downloads\da\archive\zomato.csv'

print("⏳ Reading CSV... please wait.")
# Added low_memory=False and 'c' engine for speed
df = pd.read_csv(file_path, low_memory=False, on_bad_lines='skip') 
print(f"✅ CSV Loaded! Rows found: {len(df)}")

# 2. MySQL Connection (Double check your password!)
USER = 'root'
PASSWORD = '2492005' # <-- Put your REAL password here
HOST = '127.0.0.1'
PORT = '3306'
DB_NAME = 'zomato_db'

# 3. Create Engine
engine = create_engine(f'mysql+pymysql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DB_NAME}')

# 4. Optimized Push (Chunking)
try:
    print("⏳ Pushing to MySQL with upgraded schema...")
    # 'append' use karo kyunki humne ALTER TABLE se structure badal diya hai
    df.to_sql('zomato_raw', con=engine, if_exists='append', index=False, chunksize=2000)
    print("🚀 BINGO! Finally, data is in MySQL.")
except Exception as e:
    print(f"❌ Error: {e}")
```
